```python
import matplotlib.pyplot as plt
```

```python
def f(x: float)->float:
  return (x+3)**2

def df_dx(x: float)->float:
  return 2*(x+3)
```

```python
curve = []

x_0 = float(input("Enter initial value for 'x' (x_0: )"))
n = int(input("Enter maximum number of steps to reach minima: "))
step_size = float(input("Enter step size:"))
curve.append((x_0, f(x_0)))
```

```python
x_t = x_0
for i in range(n):
  grad = df_dx(x_t)
  x_t = x_t - step_size*grad
  print(f"At step {i+1}, f(x) = {f(x_t)} and df/dx = {grad} and x = {x_t}")
  curve.append((x_t, f(x_t)))
```

```python
x, y = zip(*curve)
plt.scatter(x, y)
plt.show()
```
